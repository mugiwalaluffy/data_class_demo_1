```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# ================================================================================
# Prepare data
# ================================================================================

# Load the data
data_dir = os.path.abspath("../data/")
data_file = os.path.abspath(data_dir + "/sales_2012_2014.csv")
df = pd.read_csv(data_file)

# Show how the first few rows of the data frame look like
print(df.head())
```

```python
secname = "Revenue per retailer country from 2012 to 2014."

table = pd.pivot_table(df, values='Revenue', index=['Retailer country'], aggfunc=np.sum)
print(table)

plt.rcParams.update({'font.size': 24})
ax = table.plot.barh(subplots='True', figsize=(22,22))
plt.title(secname)
plt.show()
ax[0].get_figure().savefig('../images/revenue_country.png', bbox_inches='tight')
plt.close()

```

```python
secname = "Revenue per retailer type from 2012 to 2014."

table = pd.pivot_table(df, values='Revenue', index=['Retailer type'], aggfunc=np.sum)
print(table)

plt.rcParams.update({'font.size': 18})
ax = table.plot.pie(subplots='True', figsize=(22,22))
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.01),  shadow=True, ncol=2)
plt.title(secname)
plt.show()
ax[0].get_figure().savefig('../images/revenue_retailertype.png', bbox_inches='tight')
plt.close()


```

```python
secname = "Revenue per order method type from 2012 to 2014."

table = pd.pivot_table(df, values='Revenue', index=['Order method type'], aggfunc=np.sum)
print(table)

plt.rcParams.update({'font.size': 24})
ax = table.plot.pie(subplots='True', figsize=(12,12))
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.01),  shadow=True, ncol=2)
plt.title(secname)
plt.show()
ax[0].get_figure().savefig('../images/revenue_ordermethodtype.png', bbox_inches='tight')
plt.close()


```

```python
secname = "Revenue per product type in 2012"

df1 = df[df['Year'] == 2012]
table = pd.pivot_table(df1, values='Revenue', index=['Product type'], aggfunc=np.sum)
print(table)

plt.rcParams.update({'font.size': 24})
ax = table.plot.barh(figsize=(12,12))
plt.title(secname)
plt.show()
ax[0].get_figure().savefig('../images/revenue_producttype_2012.png', bbox_inches='tight')
plt.close()

```
